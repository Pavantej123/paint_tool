```python
import cv2
import numpy as np
import os
```

```python
x1 = None
y1 = None
draw = False
tool = "pen"
# this function handle the mouse events
def action(event,x,y,flag,user):
    global x1,y1,draw,tool
    att = color()
    if event == 1:
        x1=x
        y1=y
        draw=True
        if tool == "pen":
            cv2.circle(canva,(x,y),att[1]//2,att[0],-1)

        elif tool == "eraser":
            cv2.circle(canva,(x,y),att[1],(255, 255, 255),-1)
    elif event == 0 and draw == True:
        if tool == "pen":
            cv2.line(canva,(x1, y1),(x, y),att[0],att[1])
            x1,y1 = x,y
        elif tool == "eraser":
            cv2.line(canva,(x1, y1),(x, y),(255, 255, 255),att[1])
            x1,y1 = x,y
        
    elif event == 4:
        if tool == "line":
            cv2.line(canva,(x1,y1),(x,y),att[0],att[1])
        elif tool == "rectangle":
            cv2.rectangle(canva,(x1,y1),(x,y),att[0],att[1])
        elif tool == "circle":
            radius = int(((x-x1)**2 + (y-y1)**2)**0.5)
            cv2.circle(canva,(x1,y1),radius,att[0],att[1])
        x1=None
        y1=None
        draw=False

 #this function is for trackbars and thickness of the brush       
def color():
    b = cv2.getTrackbarPos("B","paint")
    g = cv2.getTrackbarPos("G","paint")
    r = cv2.getTrackbarPos("R","paint")
    t = cv2.getTrackbarPos("Size","paint")
    return [(b,g,r),t]
def dummy(val):
    pass

canva = np.full((1000,1000,3),255,np.uint8) #empty canvas(sheet)
cv2.namedWindow("paint",cv2.WINDOW_FREERATIO)
cv2.setMouseCallback("paint",action)
cv2.createTrackbar("R","paint",0,255,dummy)
cv2.createTrackbar("G","paint",0,255,dummy)
cv2.createTrackbar("B","paint",0,255,dummy)
cv2.createTrackbar("Size","paint",1,30,dummy)
#loop to ref
count=0
while True:
    cv2.putText(canva, "p: Pen  l: Line  r: Rectangle  c: Circle  e: Eraser",(10,25), cv2.FONT_HERSHEY_TRIPLEX,1,(0,0,0),2)
    cv2.putText(canva, "s: Save  q: Clear  a: Exit",(10,55), cv2.FONT_HERSHEY_TRIPLEX,1,(0,0,0),2)
    cv2.imshow("paint",canva)
    button = cv2.waitKey(1)
    if button == ord("a"):
        break
    if button == ord("e"):
        tool="eraser"
    if button == ord("l"):
        tool="line"
    if button == ord("r"):
        tool="rectangle"
    if button == ord("c"):
        tool="circle"
    if button == ord("p"):
        tool="pen"
    if button == ord("q"):
        canva[:]=255
    if button == ord("s"):
        count+=1
        cv2.imwrite(r"D:\Innomatics\ml\paint\img00{}.png".format(count),canva)
        print("img saved")
cv2.destroyAllWindows()
```
